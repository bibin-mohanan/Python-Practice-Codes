```python
import cv2
import mediapipe as mp
import numpy as np

mp_pose = mp.solutions.pose
pose = mp_pose.Pose()
mp_drawing = mp.solutions.drawing_utils

cap = cv2.VideoCapture(0)  
landmark_index = mp_pose.PoseLandmark.LEFT_HIP.value 
y_positions = []
frame_count = 0

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(image)

    if results.pose_landmarks:
        h, w, _ = frame.shape
        landmark = results.pose_landmarks.landmark[landmark_index]
        y_pixel = int(landmark.y * h)
        y_positions.append(y_pixel)
        frame_count += 1

      
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        cv2.circle(frame, (int(landmark.x * w), y_pixel), 5, (0, 255, 0), -1)

    cv2.imshow('Jump Height Estimator', frame)

    if cv2.waitKey(5) & 0xFF == 27:
        break
cap.release()
cv2.destroyAllWindows()

if y_positions:
    max_y = max(y_positions) 
    min_y = min(y_positions)  
    jump_height_pixels = max_y - min_y
    print(f"Estimated jump height: {jump_height_pixels} pixels")


```
